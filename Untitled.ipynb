```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn import metrics
import numpy as np

from sklearn.tree import DecisionTreeClassifier


clf=DecisionTreeClassifier()
clf.fit(X_train,y_train)
cv_scores = cross_val_score(clf, X_train, y_train, cv=5)
print(" Validatoin Scores: ",sum(cv_scores)/len(cv_scores))

y_pred=clf.predict(X_test)

# for j in X_test:
#     y_pred.append(clf.predict([j]))

acc=metrics.accuracy_score(y_test, y_pred)

print(acc)



displayConfusionMatrix(y_test,y_pred)
```

## ADABoost

```python
from sklearn.ensemble import AdaBoostClassifier
abc = AdaBoostClassifier(n_estimators=50,
                         learning_rate=1)
# Train Adaboost Classifer
model = abc.fit(X_train, y_train)

cv_scores = cross_val_score(model, X_train, y_train, cv=5)
print(" Validatoin Scores: ",sum(cv_scores)/len(cv_scores))

#Predict the response for test dataset
y_pred = model.predict(X_test)
acc=metrics.accuracy_score(y_test, y_pred)
print(acc)
print(metrics.classification_report(y_test, y_pred))

displayConfusionMatrix(y_test,y_pred)
```

## Multi Layer Perceptron

```python
from sklearn.neural_network import MLPClassifier
species=[0,1,7]
nn = MLPClassifier(random_state=1, max_iter=24,learning_rate="adaptive",verbose=1)
nn.fit(X_train, y_train)

cv_scores = cross_val_score(nn, X_train, y_train, cv=5)
print(" Validatoin Scores: ",sum(cv_scores)/len(cv_scores))

y_pred=nn.predict(X_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

print(metrics.classification_report(y_test, y_pred))
displayConfusionMatrix(y_test,y_pred)
```
