## Creating Directories for each class

```python
import os
species=list(df['Species'].unique())
for i in species:
    os.mkdir("./"+i)
```

```python
mappings={}
species=['Chinee apple', 'Lantana', 'Snake weed', 'Parthenium']
for i in species:
    mappings[i]=df[df["Species"]==i]["Filename"]
```

## moving files to respective folders

```python
import shutil
for i in mappings.keys():
    for j in mappings[i]:
        shutil.copyfile("./Weed-4class-2/"+j,"./"+i+"/"+j)
```

```python
map_vectors={}

for i in mappings.keys():
    tmp=[]
    for j in mappings[i]:
        image = skimage.io.imread(fname="./Weed-4class-2/"+j, as_gray=True)
        histogram, bin_edges = np.histogram(image, bins=256, range=(0, 1))
        tmp.append(histogram)
    map_vectors[i]=tmp
```

```python
from sklearn.model_selection import train_test_split

train_map={}
test_map={}

for i in map_vectors.keys():
    train_map[i],test_map[i],y_test,y_train=train_test_split(map_vectors[i],df[df["Species"]==i],test_size=0.3)
```

```python
map_vectors.keys()
```

```python
## making train and test variables
y_train=[]
x_train=[]
for i in train_map.keys():
    for j in train_map[i]:
        y_train.append(i)
        x_train.append(j)

y_test=[]
x_test=[]
for i in test_map.keys():
    for j in test_map[i]:
        x_test.append(j)
        y_test.append(i)
```

```python
print("trainx",len(x_train))
print("trainx",len(y_train))
print("testx",len(x_test))
print("testx",len(x_test))
```

## Training KNN

```python

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn import metrics
import numpy as np

neighbors=[1, 3, 5, 7]
val_accuracy=[]
clf=[]
mean_map={}
for n in neighbors:
    knn=KNeighborsClassifier(n_neighbors=n)
    
    clf.append(knn.fit(x_train,y_train))
    cv_scores = cross_val_score(knn, x_train, y_train, cv=5)
    cv_mean=np.mean(cv_scores)
    mean_map[n]=cv_mean
    val_accuracy.append(cv_mean)
    print("for n=",n)
    print("cv_scores mean:{}".format(cv_mean))


y_pred=[]

high_acc_n=sorted(mean_map.items(),key=lambda item:item[1],reverse=True)[0][0]

for i in range(len(neighbors)):
    if neighbors[i]==high_acc_n:
        high_acc_clf=clf[i]
        break

for j in x_test:
    y_pred.append(high_acc_clf.predict([j]))

acc=metrics.accuracy_score(y_test, y_pred)

print("Test Accuracy of highest val accuracy model(neighbors={}) is:{}".format(high_acc_n,acc))


plt.plot(neighbors,val_accuracy)

```

## NaiveBayes, Random Forest and Neural Network

### NaiveBayes

```python
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

#Train the model using the training sets
gnb.fit(x_train, y_train)
cv_scores = cross_val_score(gnb, x_train, y_train, cv=5)
cv_mean=np.mean(cv_scores)
print("cv scores",cv_scores)
print("cross-validation mean score:",cv_mean)
#Predict the response for test dataset
y_pred = gnb.predict(x_test)
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test,y_pred,labels=species))
cm=confusion_matrix(y_test,y_pred,labels=species)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=gnb.classes_)
disp.plot()
print(metrics.classification_report(y_test, y_pred))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(x_train,y_train)
cv_scores = cross_val_score(clf, x_train, y_train, cv=5)
cv_mean=np.mean(cv_scores)
print("cv scores",cv_scores)
print("cross-validation mean score:",cv_mean)
#Predict the response for test dataset
y_pred = clf.predict(x_test)
print(y_pred)
print("len(y_pred)=",len(y_pred))
print("len(x_test)=",len(x_test))
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
cm=confusion_matrix(y_test,y_pred,labels=species)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=gnb.classes_)
disp.plot()

print(metrics.classification_report(y_test, y_pred))
```

### Multi Layer Perceptron

```python
from sklearn.neural_network import MLPClassifier

species=['Chinee apple', 'Lantana', 'Snake weed', 'Parthenium']
clf = MLPClassifier(random_state=1, max_iter=300,learning_rate="adaptive").fit(x_train, y_train)

cv_scores = cross_val_score(clf, x_train, y_train, cv=5)
cv_mean=np.mean(cv_scores)
print("cv scores",cv_scores)
print("cross-validation mean score:",cv_mean)

y_pred=clf.predict(x_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

cm=confusion_matrix(y_test,y_pred,labels=species)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=gnb.classes_)
disp.plot()


print(metrics.classification_report(y_test, y_pred))
```
